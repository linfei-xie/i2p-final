# Executive Briefing

## Review of the evidence

The rapid development of the Airbnb platform has attracted the attention of the government. They believe that the Airbnb platform "enables Londoners to participate in the sharing economy and benefit from recent innovations in information technology by renting out an empty room or entire house in the same way as other residents across the country." "(Department for Communities and Local Government, 2015), so in the "Deregulation Act" passed in March 2015, the government proposed and implemented measures to eliminate "red tape", allowing less than 90 days in a calendar year Short-term rental without changing the use (Ferreri and & Sanyal, 2018), which is also the main basis for the current London government to manage the Airbnb platform, and there are no clear requirements for the operation of the business team. 

The involvement of the commercialization team within the Airbnb platform poses new challenges to the platform and the hotel industry, because real interpersonal contact is one of the key drivers for Airbnb to obtain memorable experiences and brand loyalty (Mody, Hanks & Dograu, 2019) In the long run, this value challenge may be very difficult for the platform. For the hotel industry, with the professional operation of Airbnb, its management advantage is weakening, which intensifies the competition between Airbnb and hotels. In addition, compared with traditional hotels, Airbnb is easier to spread in cities, because the establishment of a hotel requires the entire building to be available, and the required land also requires relevant government permits, while Airbnb can expand anywhere in existing houses and apartment buildings Supply (Guttentag, D, 2015), which puts the hotel industry at a disadvantage compared to Airbnb housing management companies.  

The profitability of individual operators and professional operators on the Airbnb platform is different. A study by Paschalis et al. (2020) shows that there is a clear gap between the pricing of rental housing between individual operators in Greece and professional operators with multiple listings. This price gap is attributed to the good management and operation methods of professional operators. In the New York Airbnb market, commercial operators that control multiple whole-house/apartment listings or large private room combinations account for only 12% of the host, but their revenue in New York City exceeds 28% (Wachsmuth D, et al. 2018).

**Airbnb in London**

1.Distribution:  

As of August 24, 2020, there are 52175 houses on Airbnb platform in London(after data cleaning). The number of houses in each district of London is shown in the figure below. According to the picture, the distribution shows Westminster as the core and spreading to the surrounding areas. Many houses are concentrated in the center of the city and around tourist attractions (London eye, tower of London, British Museum, etc.), while there are few houses in the peripheral areas of the city. This analysis confirms the conclusion of Xu et al. (2020) on the factors influencing the location of Airbnb houses in London, and the study also points out that Airbnb houses in London are concentrated in the city center and around tourist attractions, and are related to the cultural and art spots, Waterscape, vegetation, transportation facilities, nightlife and other factors around accommodation.

2.Price:  

The average price of Airbnb listings in London is 110£/d, which is down from the average price of 126£/d at the beginning of 2020(inside-Airbnb). The average price of housing in various districts of London also shows the overall characteristics of high center and low periphery. Different from the distribution of quantity, the area with the highest price is the City of London, with an average unit price of 220£/d, the second highest rent, and the third The high areas are Kensington and Chelsea and Westminster in the West End of London. The lowest average unit price is located in the Sutton area of South London, at 53£/d.

3.Houses type:  

London Airbnb offers four types of housing: Entire home/apartment, Hotel room, Private room, and Shared room. The type of housing for the whole rental is the most, accounting for 57.62% of the total housing. Private rooms and shared rooms account for them respectively. 41.01% of the total listings, 0.54%, there are still hotel room rentals on the Airbnb platform, accounting for 0.82% of the total listings.

4.Operation:
    
Based on the existing data, this report provides a rough estimate of the occupancy rate (‘average monthly reviews’*‘minimum stay’/30), nights per year (‘average monthly review’*’minimum stay’/30*365) and each listing average monthly income (‘monthly average number of reviews’ * ‘minimum stay days’ * ‘price’) these three indicators help readers better understand the operation of the London Airbnb platform. According to calculations, as of August 24, 2020, the average occupancy rate of the Airbnb platform is 8.2%, the average annual night’s stay is 30, and the average monthly income from listings is 294 £/d. On the inside-Airbnb website, we found the data on January 9, 2020. The data shows that the average occupancy rate of the Airbnb platform during this period was 23.2%, the average annual night’s stay was 85, and the average monthly income from listings was £762/ d. The time interval between the two statistics is only 8 months. The data gap is so large. Apart from statistical errors, the main reason should be the global Covid-19 epidemic. Since the outbreak of the Covid-19 in the UK in March, many travels to the UK has been canceled. The reduction in travel activities has had a huge impact on the Airbnb platform business. The operating conditions in different districts of London are also different. The area with the highest occupancy rate and average number of nights per year is the City of London, and the area with the highest monthly average income is Haringey, 1,564£/m.

**London Airbnb hosts**

1.Overall:
    
There are 52,175 listings on the London Airbnb platform, with 35,123 hosts. Each host manages 1.49 homes on average. The host who manages the most listings is called Alex, with 363 houses, which is a typical business operation team on Airbnb platform. According to the number of houses available, all hosts are divided into four categories: individual business host managing 1-2 suites, small business host managing 3-10 suites, middle business host managing 11-50 suites and big business host managing more than 50 suites. The latter three categories can be regarded as business teams operating Airbnb. According to statistics, most of houses are managed by individuals, accounting for 65.57% of the total, followed by small companies (19.33%), followed by medium-sized companies (10.09%), and only a few houses (4.92%) are managed by large companies. 2.Distribution:
    
The houses managed by the four types of hosts with different management scales generally have a similar distribution pattern of "more centers and fewer outsiders". Individual hosts have the most housing in the Hackney area. The other three types of professional landlords are surprisingly consistent in choosing housing locations. The most housing is located in the Westminster area. In addition, big business hosts have no houses in Kingston upon Thames, Sutton in South London, Bexley, Barking and Dagenham in East London. The location of individual hosts and business teams can reflect different business strategies to a certain extent. Individual hosts mostly use their own houses as the source of houses, so they are more widely distributed. Hackney District, which has the largest number of houses, is closer to the city center and has convenient transportation, but compared with the Western District, the house price is lower and the operation risk is lower. Business operation teams tend to manage and operate in the city Heart, Westminster District, with the most abundant tourism resources around, is the most preferred area for tourists to live in, which can improve the occupancy rate, but the high housing price also requires its better operation ability. In addition, the largest management team may consider the income and other issues and will not manage the low monthly income areas.
